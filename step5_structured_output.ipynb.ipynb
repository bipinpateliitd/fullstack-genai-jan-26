```python
from typing import TypedDict, Annotated

class Person(TypedDict):
    name: Annotated[str, "Full name of the person"]
    age: Annotated[int, "Age in completed years"]
    profession: Annotated[str, "Primary occupation"]
```

```python
class FIRSummary(TypedDict):
    complainant_name: Annotated[
        str,
        "Full name of the complainant"
    ]
    incident_type: Annotated[
        str,
        "Short label like Theft, Assault, Fraud"
    ]
    location: Annotated[
        str,
        "Place of incident including city"
    ]
    severity: Annotated[
        str,
        "One of: Low, Medium, High"
    ]
```
