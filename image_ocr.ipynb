```python
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from typing import TypedDict, Literal
from langchain_core.messages import HumanMessage
import base64
from pathlib import Path

load_dotenv(dotenv_path="/home/bipin/Documents/genai/g25-nov-hindi/langchain-learning/.env")
llm = ChatOpenAI(
        model="gpt-4.1-nano",
        temperature=0.7,
        max_tokens=200,
    )
```

```python
class Id(TypedDict):
    name:str
    age: int
    dob: str
    year: int
    address: str
    phone_number: int
    email_id: str
    university_name: str
    school_name: str
```

```python
str_llm = llm.with_structured_output(Id)
```

```python
# Method 2: Local image as base64
def encode_image(image_path: str) -> str:
    """Encode local image to base64 data URI."""
    with open(image_path, "rb") as f:
        image_data = base64.b64encode(f.read()).decode("utf-8")
    return f"data:image/jpeg;base64,{image_data}"

image_path = Path("s1.png")
base64_image = encode_image(image_path)
```

```python
message = HumanMessage(
    content=[
        {"type": "text", "text": "Extract the text."},
        {
            "type": "image_url",
            "image_url": {"url": base64_image},
        },
    ]
)
response = str_llm.invoke([message])
response
```

```python

# Create a message with image URL
image_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/d/dd/Gfp-wisconsin-madison-the-nature-boardwalk.jpg/2560px-Gfp-wisconsin-madison-the-nature-boardwalk.jpg"

# Using content blocks (LangChain standard format)
message = HumanMessage(
    content=[
        {"type": "text", "text": "What's in this image? Describe it in short in Hindi."},
        {"type": "image_url", "image_url": {"url": image_url}}
    ]
)

response = llm.invoke([message])
print(response.content)
```
